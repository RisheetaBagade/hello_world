```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
my_movies=pd.read_csv("my_movies.csv")
my_movies
```

```python
my_movies = my_movies.replace(np.nan, '', regex=True)
my_movies 
```

```python
my_movies.shape
```

```python
my_movies
viewed = []
for i in range(0, 10):
    viewed.append([str(my_movies.values[i,u]) for u in range(1, 15)])
```

```python
my_movies_series  = pd.DataFrame(pd.Series(viewed))
my_movies_series 
my_movies_series.columns = ["MovieViews"]
my_movies_series
```

```python
X = my_movies_series['MovieViews'].str.join(sep=',').str.get_dummies(sep=',')
X
```

```python
X.drop(["0"],inplace=True,axis=1)
X
```

```python
X.drop(["1"],inplace=True,axis=1)
X
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
frequent_itemsets = apriori(X, min_support=0.005, max_len=3,use_colnames = True)
frequent_itemsets
```

```python
frequent_itemsets.sort_values('support',ascending = False,inplace=True)
frequent_itemsets.sort_values
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(20)
```

```python
rules.sort_values('lift',ascending = False).head(10)
```
