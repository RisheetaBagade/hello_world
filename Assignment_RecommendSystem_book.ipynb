```python
#User Based Collaborative Filtering
import pandas as pd
import numpy as np
```

```python
book=pd.read_csv("book1.csv",encoding = "ISO-8859-1")
book
```

```python
book.columns
```

```python
book =book.rename(columns={"User.ID":"Userid","Book.Title":"Booktitle","Book.Rating":"Bookrating"})
book.head()
```

```python
#number of unique users in the dataset
len(book.Userid.unique())
```

```python
#number of unique users in the dataset
len(book.Booktitle.unique())
```

```python
book.drop_duplicates(subset="Userid", keep=False, inplace=True)
```

```python
book=book.drop_duplicates(subset=["Booktitle"])
```

```python
book.shape
```

```python
book_df = book.pivot(index='Userid',
                                 columns='Booktitle',
                                 values='Bookrating').reset_index(drop=True)
```

```python
book_df.iloc[500:600,500:600]
```

```python
book_df.index = book.Userid.unique()
```

```python
book_df
```

```python
#Impute those NaNs with 0 values
book_df.fillna(0, inplace=True) 
```

```python
book_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( book_df.values,metric='cosine') 
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book.Userid.unique()
user_sim_df.columns = book.Userid.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```

```python
#fetching ratings
Married_user_rating=book_df['Married In Haste (Intrigue , No 481)']
```

```python
Married_user_rating
```

```python
#Finding the correlation with different movies
similar_to_Married = book_df.corrwith(Married_user_rating) 
```

```python
similar_to_Married.head(10) 
```

```python
corr_Married = pd.DataFrame(similar_to_Married, columns=['Correlation'])
corr_Married.dropna(inplace=True)
corr_Married.head()
```

```python
corr_Married[corr_Married['Correlation'] > 0].sort_values(by='Correlation', ascending=False).head(5)  
```
