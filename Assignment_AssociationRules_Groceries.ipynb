```python
import pandas as pd 
import seaborn as sb 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
groceries = []
with open("groceries.csv") as f:
    groceries = f.read()
groceries = groceries.split("\n")
groceries
groceries_list = []
for i in groceries:
    groceries_list.append(i.split(","))
groceries_list
all_groceries_list = [i for item in groceries_list for i in item] #for pull out all item from each row and present in column wise
all_groceries_list 
```

```python
from collections import Counter,OrderedDict
```

```python
item_frequencies = Counter(all_groceries_list)
item_frequencies
```

```python
# after sorting
item_frequencies = sorted(item_frequencies.items(),key = lambda x:x[1])
item_frequencies
```

```python
frequencies = list(reversed([i[1] for i in item_frequencies]))
frequencies
```

```python
items = list(reversed([i[0] for i in item_frequencies]))
items
```

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
import matplotlib.pyplot as plt
plt.bar(x=list(range(0,5)) ,height=frequencies[0:5],color='rgbkymc')
plt.xticks(list(range(0,5),),items[0:5])
plt.xlabel("items")
plt.ylabel("Count")
plt.xlabel("Items")
plt.title('Top 5 items')
```

```python
groceries_series  = pd.DataFrame(pd.Series(groceries_list))
groceries_series = groceries_series.iloc[:9835,:] # removing the last empty transaction
groceries_series
groceries_series.columns = ["transactions"]
groceries_series
```

```python
X = groceries_series['transactions'].str.join(sep='*').str.get_dummies(sep='*')
X
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
frequent_itemsets = apriori(X, min_support=0.006, max_len=3,use_colnames = True)
frequent_itemsets
```

```python
frequent_itemsets.sort_values('support',ascending = False,inplace=True)
frequent_itemsets.sort_values
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(20)
rules.sort_values('lift',ascending = False).head(10)
```
